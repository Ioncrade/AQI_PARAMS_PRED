```python
import pandas as pd 
import numpy as np 
```

```python
df_1= pd.read_csv('P_2023.csv')
df_2 = pd.read_csv('P_2022.csv')
df_3 = pd.read_csv('P_2021.csv')
df_4 = pd.read_csv('P_2020.csv')
df_5 = pd.read_csv('P_2019.csv')
df_6 = pd.read_csv('P_2018.csv')
df =  pd.concat([df_6,df_5,df_4,df_3,df_2,df_1],ignore_index = True)
df['Timestamp'] = pd.to_datetime(df['Timestamp']) #Changing obj to datetime(ns) format
df.set_index('Timestamp', inplace=True)
```

```python
df_pm = pd.DataFrame(columns=['PM2.5 (µg/m³)'])
df_pm['PM2.5 (µg/m³)']= df['PM2.5 (µg/m³)']

df_pm10 = pd.DataFrame(columns=['PM10 (µg/m³)'])
df_pm10['PM10 (µg/m³)'] = df['PM10 (µg/m³)']

df_no2 = pd.DataFrame(columns=['NO2 (µg/m³)'])
df_no2['NO2 (µg/m³)'] = df['NO2 (µg/m³)']

df_co = pd.DataFrame(columns=['CO (mg/m³)'])
df_co['CO (mg/m³)'] = df['CO (mg/m³)']

df_so2 = pd.DataFrame(columns=['SO2 (µg/m³)'])
df_so2['SO2 (µg/m³)'] = df['SO2 (µg/m³)']
```

```python
import matplotlib.pyplot as plt

# Plot for PM2.5
plt.figure(figsize=(14, 6))
plt.plot(df_pm.index, df_pm['PM2.5 (µg/m³)'], label='PM2.5 (µg/m³)', color='blue')
plt.title('PM2.5 (µg/m³) Over Time')
plt.xlabel('Timestamp')
plt.ylabel('Concentration')
plt.legend()
plt.show()

# Plot for PM10
plt.figure(figsize=(14, 6))
plt.plot(df_pm10.index, df_pm10['PM10 (µg/m³)'], label='PM10 (µg/m³)', color='green')
plt.title('PM10 (µg/m³) Over Time')
plt.xlabel('Timestamp')
plt.ylabel('Concentration')
plt.legend()
plt.show()

# Plot for NO2
plt.figure(figsize=(14, 6))
plt.plot(df_no2.index, df_no2['NO2 (µg/m³)'], label='NO2 (µg/m³)', color='red')
plt.title('NO2 (µg/m³) Over Time')
plt.xlabel('Timestamp')
plt.ylabel('Concentration')
plt.legend()
plt.show()

# Plot for CO
plt.figure(figsize=(14, 6))
plt.plot(df_co.index, df_co['CO (mg/m³)'], label='CO (mg/m³)', color='purple')
plt.title('CO (mg/m³) Over Time')
plt.xlabel('Timestamp')
plt.ylabel('Concentration')
plt.legend()
plt.show()

# Plot for SO2
plt.figure(figsize=(14, 6))
plt.plot(df_so2.index, df_so2['SO2 (µg/m³)'], label='SO2 (µg/m³)', color='orange')
plt.title('SO2 (µg/m³) Over Time')
plt.xlabel('Timestamp')
plt.ylabel('Concentration')
plt.legend()
plt.show()
```

```python
from pandas.plotting import lag_plot
import matplotlib.pyplot as plt

# PM2.5 Lag Plots
fig, ax_ = plt.subplots(1, 4, figsize=(25, 5))
pm_series = df_pm['PM2.5 (µg/m³)']
for i in range(4):
    lag_plot(pm_series, lag=i+1, ax=ax_[i])
    ax_[i].plot([min(pm_series), max(pm_series)], [min(pm_series), max(pm_series)], '--', color='red')
fig.suptitle('Lag Plots for PM2.5 (µg/m³)')
plt.show()

# PM10 Lag Plots
fig, ax_ = plt.subplots(1, 4, figsize=(25, 5))
pm10_series = df_pm10['PM10 (µg/m³)']
for i in range(4):
    lag_plot(pm10_series, lag=i+1, ax=ax_[i])
    ax_[i].plot([min(pm10_series), max(pm10_series)], [min(pm10_series), max(pm10_series)], '--', color='red')
fig.suptitle('Lag Plots for PM10 (µg/m³)')
plt.show()

# NO2 Lag Plots
fig, ax_ = plt.subplots(1, 4, figsize=(25, 5))
no2_series = df_no2['NO2 (µg/m³)']
for i in range(4):
    lag_plot(no2_series, lag=i+1, ax=ax_[i])
    ax_[i].plot([min(no2_series), max(no2_series)], [min(no2_series), max(no2_series)], '--', color='red')
fig.suptitle('Lag Plots for NO2 (µg/m³)')
plt.show()

# CO Lag Plots
fig, ax_ = plt.subplots(1, 4, figsize=(25, 5))
co_series = df_co['CO (mg/m³)']
for i in range(4):
    lag_plot(co_series, lag=i+1, ax=ax_[i])
    ax_[i].plot([min(co_series), max(co_series)], [min(co_series), max(co_series)], '--', color='red')
fig.suptitle('Lag Plots for CO (mg/m³)')
plt.show()

# SO2 Lag Plots
fig, ax_ = plt.subplots(1, 4, figsize=(25, 5))
so2_series = df_so2['SO2 (µg/m³)']
for i in range(4):
    lag_plot(so2_series, lag=i+1, ax=ax_[i])
    ax_[i].plot([min(so2_series), max(so2_series)], [min(so2_series), max(so2_series)], '--', color='red')
fig.suptitle('Lag Plots for SO2 (µg/m³)')
plt.show()
```

```python
# PM2.5
df_pm['PM2.5 (µg/m³)'] = pd.to_numeric(df_pm['PM2.5 (µg/m³)'], errors='coerce')
mean_pm25 = df_pm['PM2.5 (µg/m³)'].mean()
df_pm['PM2.5 (µg/m³)'] = df_pm['PM2.5 (µg/m³)'].fillna(mean_pm25)

# PM10
df_pm10['PM10 (µg/m³)'] = pd.to_numeric(df_pm10['PM10 (µg/m³)'], errors='coerce')
mean_pm10 = df_pm10['PM10 (µg/m³)'].mean()
df_pm10['PM10 (µg/m³)'] = df_pm10['PM10 (µg/m³)'].fillna(mean_pm10)

# NO2
df_no2['NO2 (µg/m³)'] = pd.to_numeric(df_no2['NO2 (µg/m³)'], errors='coerce')
mean_no2 = df_no2['NO2 (µg/m³)'].mean()
df_no2['NO2 (µg/m³)'] = df_no2['NO2 (µg/m³)'].fillna(mean_no2)

# CO
df_co['CO (mg/m³)'] = pd.to_numeric(df_co['CO (mg/m³)'], errors='coerce')
mean_co = df_co['CO (mg/m³)'].mean()
df_co['CO (mg/m³)'] = df_co['CO (mg/m³)'].fillna(mean_co)

# SO2
df_so2['SO2 (µg/m³)'] = pd.to_numeric(df_so2['SO2 (µg/m³)'], errors='coerce')
mean_so2 = df_so2['SO2 (µg/m³)'].mean()
df_so2['SO2 (µg/m³)'] = df_so2['SO2 (µg/m³)'].fillna(mean_so2)
```

```python
# Seasonal Decompose using statsmodels library.

from statsmodels.tsa.seasonal import seasonal_decompose
import matplotlib.pyplot as plt

# PM2.5 Seasonal Decomposition
PM_data = df_pm['PM2.5 (µg/m³)']
result_pm25 = seasonal_decompose(PM_data, model='multiplicative')
plt.tight_layout()
result_pm25.plot()
plt.suptitle('PM2.5 (µg/m³)')
plt.show()

# PM10 Seasonal Decomposition
PM10_data = df_pm10['PM10 (µg/m³)']
result_pm10 = seasonal_decompose(PM10_data, model='multiplicative')
plt.tight_layout()
result_pm10.plot()
plt.suptitle('PM10 (µg/m³)')
plt.show()

# NO2 Seasonal Decomposition
NO2_data = df_no2['NO2 (µg/m³)']
result_no2 = seasonal_decompose(NO2_data, model='multiplicative')
plt.tight_layout()
result_no2.plot()
plt.suptitle('NO2 (µg/m³)')
plt.show()

# CO Seasonal Decomposition
CO_data = df_co['CO (mg/m³)']
result_co = seasonal_decompose(CO_data, model='additive')
plt.tight_layout()
result_co.plot()
plt.suptitle('CO (mg/m³)')
plt.show()

# SO2 Seasonal Decomposition
SO2_data = df_so2['SO2 (µg/m³)']
result_so2 = seasonal_decompose(SO2_data, model='multiplicative')
plt.tight_layout()
result_so2.plot()
plt.suptitle('SO2 (µg/m³)')
plt.show()
```

```python

# RESAMPLING DATA TO MONTHLY FORMAT
# MS is code for resampling Monthly

seasonal_data = PM_data
seasonal_data = seasonal_data.resample(rule='MS').mean()
seasonal_data# Resample and calculate the mean for each parameter

# PM2.5
seasonal_pm25 = df_pm['PM2.5 (µg/m³)']
seasonal_pm25 = seasonal_pm25.resample(rule='MS').mean()

# PM10
seasonal_pm10 = df_pm10['PM10 (µg/m³)']
seasonal_pm10 = seasonal_pm10.resample(rule='MS').mean()

# NO2
seasonal_no2 = df_no2['NO2 (µg/m³)']
seasonal_no2 = seasonal_no2.resample(rule='MS').mean()

# CO
seasonal_co = df_co['CO (mg/m³)']
seasonal_co = seasonal_co.resample(rule='MS').mean()

# SO2
seasonal_so2 = df_so2['SO2 (µg/m³)']
seasonal_so2 = seasonal_so2.resample(rule='MS').mean()

# Output the resampled data to verify
seasonal_pm25, seasonal_pm10, seasonal_no2, seasonal_co, seasonal_so2
```

```python
from statsmodels.graphics import tsaplots
import matplotlib.pyplot as plt

# Calculate the number of lags to use
def plot_acf_for_parameter(data, title):
    n_observations = len(data)
    lags = min(24, n_observations - 1)
    fig = tsaplots.plot_acf(data, lags=lags)
    plt.title(title)
    plt.show()

# PM2.5
monthly_pm25 = df_pm['PM2.5 (µg/m³)'].resample(rule='MS').mean()
plot_acf_for_parameter(monthly_pm25, "Autocorrelation Plot for PM2.5 (µg/m³)")

# PM10
monthly_pm10 = df_pm10['PM10 (µg/m³)'].resample(rule='MS').mean()
plot_acf_for_parameter(monthly_pm10, "Autocorrelation Plot for PM10 (µg/m³)")

# NO2
monthly_no2 = df_no2['NO2 (µg/m³)'].resample(rule='MS').mean()
plot_acf_for_parameter(monthly_no2, "Autocorrelation Plot for NO2 (µg/m³)")

# CO
monthly_co = df_co['CO (mg/m³)'].resample(rule='MS').mean()
plot_acf_for_parameter(monthly_co, "Autocorrelation Plot for CO (mg/m³)")

# SO2
monthly_so2 = df_so2['SO2 (µg/m³)'].resample(rule='MS').mean()
plot_acf_for_parameter(monthly_so2, "Autocorrelation Plot for SO2 (µg/m³)")
```

```python
x_train_pm25 = monthly_pm25['2018-01-01 00:00:00':'2022-10-19 00:00:00']
x_test_pm25 = monthly_pm25['2022-10-19 01:00:00':'2023-12-31 23:00:00']

x_train_pm10 = monthly_pm10['2018-01-01 00:00:00':'2022-10-19 00:00:00']
x_test_pm10 = monthly_pm10['2022-10-19 01:00:00':'2023-12-31 23:00:00']

x_train_no2 = monthly_no2['2018-01-01 00:00:00':'2022-10-19 00:00:00']
x_test_no2 = monthly_no2['2022-10-19 01:00:00':'2023-12-31 23:00:00']

x_train_co = monthly_co['2018-01-01 00:00:00':'2022-10-19 00:00:00']
x_test_co = monthly_co['2022-10-19 01:00:00':'2023-12-31 23:00:00']

x_train_so2 = monthly_so2['2018-01-01 00:00:00':'2022-10-19 00:00:00']
x_test_so2 = monthly_so2['2022-10-19 01:00:00':'2023-12-31 23:00:00']
```

```python
model_pm25=SARIMAX(x_train_pm25,order=(1,0,0),seasonal_order=(1,0,1,12),)
results=model_pm25.fit()

predictions_pm25 = results.predict(start=list(x_test_pm25.index)[0], end=list(x_test_pm25.index)[-1], typ='levels').rename('Predictions')
```

```python
model_pm10 = SARIMAX(x_train_pm10,order=(1,0,0),seasonal_order=(1,0,1,12),)
results=model_pm10.fit()

predictions_pm10 = results.predict(start=list(x_test_pm10.index)[0], end=list(x_test_pm10.index)[-1], typ='levels').rename('Predictions')
```

```python
model_co = SARIMAX(x_train_co,order=(1,0,0),seasonal_order=(1,0,1,12),)
results=model_co.fit()

predictions_co = results.predict(start=list(x_test_co.index)[0], end=list(x_test_co.index)[-1], typ='levels').rename('Predictions')
```

```python
model_no2 = SARIMAX(x_train_no2,order=(2,0,0),seasonal_order=(1,0,0,11),)
results=model_no2.fit()

predictions_no2 = results.predict(start=list(x_test_no2.index)[0], end=list(x_test_no2.index)[-1], typ='levels').rename('Predictions')
```

```python
model_so2 = SARIMAX(x_train_so2,order=(1,0,0),seasonal_order=(1,0,1,12),)
results=model_so2.fit()

predictions_so2 = results.predict(start=list(x_test_so2.index)[0], end=list(x_test_so2.index)[-1], typ='levels').rename('Predictions')
```

```python
import numpy as np
import math
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score, explained_variance_score, max_error, mean_poisson_deviance, mean_gamma_deviance

# Assuming you already have your predictions and test data for each parameter
parameters = {
    'PM2.5': (x_test_pm25, predictions_pm25),
    'PM10': (x_test_pm10, predictions_pm10),
    'NO2': (x_test_no2, predictions_no2),
    'CO': (x_test_co, predictions_co),
    'SO2': (x_test_so2, predictions_so2)
}

# Iterate through each parameter and calculate the metrics
for param, (x_test, predictions) in parameters.items():
    print(f"Metrics for {param}:")
    
    rmse = np.sqrt(mean_squared_error(x_test, predictions))
    print(f'  Root Mean Squared Error (RMSE): {rmse}')
    
    mse = mean_squared_error(x_test, predictions)
    print(f'  Mean Squared Error (MSE): {mse}')
    
    mae = mean_absolute_error(x_test, predictions)
    print(f'  Mean Absolute Error (MAE): {mae}')
    
    mape = np.mean(np.abs(predictions - x_test) / np.abs(x_test))
    print(f'  Mean Absolute Percentage Error (MAPE): {mape}')
    
    r2score = r2_score(x_test, predictions)
    print(f'  R² Score: {r2score}')
    
    explained_variance = explained_variance_score(x_test, predictions)
    print(f'  Explained Variance Score: {explained_variance}')
    
    me = max_error(x_test, predictions)
    print(f'  Max Error: {me}')
    
    mpd = mean_poisson_deviance(x_test, predictions)
    print(f'  Mean Poisson Deviance: {mpd}')
    
    mgd = mean_gamma_deviance(x_test, predictions)
    print(f'  Mean Gamma Deviance: {mgd}')
    
    forecast_errors = x_test - predictions
    bias = np.mean(forecast_errors)
    print(f'  Bias: {bias}')
    
    print("\n" + "-"*50 + "\n")
```
